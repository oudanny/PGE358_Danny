```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib qt
```

```python
df = pd.read_excel('HW2_data.xlsx')
df['DEPTH'] = df['~A DEPTH ']
max_depth = df['DEPTH'].max()
df = df.drop('~A DEPTH ', axis='columns')
df['NPHI (ss units)'] = df['NPHI'] 
df['CALI_in'] = df['CALI']*12
df = df.loc[:1960]
```

```python
df.head()
```

#### Part I. Logs 1

```python
#THIS IS THE UPDATED CODE FOR CURRENT VERSIONS (3.9.2) OF MATPLOTLIB!!!

BH_temp = 140
Surface_temp = 40
temp_grad = (BH_temp - Surface_temp) / max_depth
df['TEMP [F]'] = Surface_temp + df['DEPTH'] * temp_grad

# Create subplots with a specific grid layout (3 columns in one row)
fig, axes = plt.subplots(1, 5, figsize=(3.5*2, 3.5*4), sharey=True)

# Curve names for labeling
curve_names = ['Caliper\n (CALI, in)', 'Gamma Ray\n (GR, API)', 'Resistivity\n (RILD, ohm.m)', 
               'Bulk Density\n (RHOB, g/cc)', 'Neutron Porosity\n (SS units)', 'Temperature [F]','DT [Î¼s/ft]']

ax1 = axes[0]
ax2 = ax1.twiny()  # Create secondary axis for Caliper plot
ax3 = axes[1]
ax4 = axes[2]
ax5 = ax4.twiny()  # Create secondary axis for Neutron Porosity plot
ax6 = axes[4]
ax7 = axes[3]

# Plot data for each axis
ax1.plot(df['CALI_in'], df['DEPTH'], color='red', linewidth=1.25)
ax1.invert_yaxis()
ax1.set_ylabel('Depth (Ft)')
ax1.set_xlim(5, 20)
ax1.xaxis.label.set_color('red')
ax1.tick_params(axis='x', colors='red')
#ax1.set_ylim(df['DEPTH'].min(), df['DEPTH'].max())  # Manually set y-axis limits

ax2.plot(df['GR'], df['DEPTH'], color='green', linewidth=1.25)
ax2.set_xlim(0, 150)
ax2.spines['top'].set_position(('axes', 1.08))
ax2.spines['top'].set_visible(True)
ax2.xaxis.label.set_color('green')
ax2.tick_params(axis='x', colors='green')

ax3.plot(df['RILD'], df['DEPTH'], color='red', linewidth=1.25)
ax3.set_xscale('log')
ax3.set_xlim(0.2, 200)
ax3.xaxis.label.set_color('red')
ax3.tick_params(axis='x', colors='red')

ax4.plot(df['RHOB'], df['DEPTH'], color='red', linewidth=1.25)
ax4.set_xlim(1.65, 2.65)
ax4.xaxis.label.set_color('red')
ax4.tick_params(axis='x', colors='red')
ax4.set_xticks([1.65,2.05, 2.25, 2.65])

ax5.plot(df['NPHI (ss units)'], df['DEPTH'], color='navy', linewidth=1.25)
ax5.set_xlim(60, 0)
ax5.spines['top'].set_position(('axes', 1.08))
ax5.spines['top'].set_visible(True)
ax5.xaxis.label.set_color('navy')
ax5.tick_params(axis='x', colors='navy')
#ax5.fill_betweenx(df['DEPTH'], df['RHOB'], df['NPHI (ss uni)'],
#                 where=(df['RHOB'] < df['NPHI (ss uni)']), color='cyan', interpolate=True)

ax6.plot(df['TEMP [F]'], df['DEPTH'],linewidth=1.25,color='#BF5700')
# ax6.spines['top'].set_position(('axes', 1.08))
ax6.spines['top'].set_visible(True)
ax6.set_xlim(130, 140)
ax6.xaxis.label.set_color('#BF5700')
ax6.tick_params(axis='x', colors='#BF5700')
ax6.set_xticks(np.linspace(130,140.1,3).astype(int))

ax7.plot(df['DT'], df['DEPTH'], color='green', linewidth=1.25)
ax7.set_xlim(80, 180)
# ax7.spines['top'].set_position(('axes', 1.08))
ax7.spines['top'].set_visible(True)
ax7.xaxis.label.set_color('green')
ax7.tick_params(axis='x', colors='green')

# Manually set the y-axis for all subplots
for ax in [ax1, ax2, ax3, ax4, ax5,ax6,ax7]:
    #ax.set_ylim(df['DEPTH'].min(), df['DEPTH'].max())  # Ensure y-axis is correctly set
    ax.xaxis.set_ticks_position('top')
    ax.xaxis.set_label_position('top')
    # Define the number of gridlines (20) for each axis
    x_min, x_max = ax.get_xlim()

    if ax != ax3:  # Logarithmic scale for Resistivity plot
        grid_positions = np.linspace(x_min, x_max, 20).astype(int)  # Linear-spaced gridlines

    elif ax == ax3:
        nothing = "True"



    if ax != ax3:
        ax.set_xticks(grid_positions, minor=True)  # Use minor ticks to avoid excessive labeling
        ax.grid(True, axis='x', linestyle='--', linewidth=0.5, which='minor')  # Gridlines on minor ticks
    elif ax == ax3:
        ax.grid(True, axis='x')   

# Use a for loop to assign x-axis labels from curve_names
axes_list = [ax1, ax2, ax3, ax4, ax5,ax6,ax7]  # List of axes
for i, ax in enumerate(axes_list):
    ax.set_xlabel(curve_names[i])  # Set the x-axis label from curve_names

# Hide y-tick labels for all axes except the first one
for ax in fig.axes[1:]:
    plt.setp(ax.get_yticklabels(), visible=False)


# Adjust spacing to prevent overlap
fig.subplots_adjust(wspace=.1, hspace=0)

# Title and finalize layout
#plt.suptitle('WR313 H001', x=0.1, horizontalalignment='left', fontsize='xx-large', fontweight='bold')

# Save plot

# Show plot
# plt.show()
plt.savefig(r'.\plots\HW_3_PartI_1_Well_logs_plot.png', format='png', dpi=300, bbox_inches='tight')
plt.close()
```

#### Part I. Logs 2

```python
# Density porosity
df['Apparent Density Porosity'] = (2.65 - df['RHOB'])/(2.65-1)

# Apparent Water resistivity 
a = 1
m = 2.05
n = 1.95
df['Apparent Water Resistivity'] = df['RILD'] * np.power(df['Apparent Density Porosity'],m)/a
```

```python
fig, axes = plt.subplots(1, 3, figsize=(3.5*2, 3.5*4), sharey=True)

# Curve names for labeling
curve_names = ['Temperature [F]', 'Apparent Density Por.\nSS Units','Neutron Porosity\n (SS units)','Apparent Water Resistivity']

ax1 = axes[0]
# ax2 = ax1.twiny()  # Create secondary axis for Caliper plot
ax2 = axes[1]
ax3 = ax2.twiny()
ax4 = axes[2]


# Plot data for each axis
ax1.plot(df['TEMP [F]'], df['DEPTH'], color='red', linewidth=1.25)
ax1.invert_yaxis()
ax1.set_ylabel('Depth (Ft)')
ax1.set_xlim(130, 140)
ax1.xaxis.label.set_color('red')
ax1.tick_params(axis='x', colors='red')
ax1.set_xticks([130,135, 140])
#ax1.set_ylim(df['DEPTH'].min(), df['DEPTH'].max())  # Manually set y-axis limits

ax2.plot(df['Apparent Density Porosity'], df['DEPTH'], color='green', linewidth=1.25)
ax2.set_xlim(0.6, 0)
ax2.spines['top'].set_position(('axes', 1.08))
ax2.spines['top'].set_visible(True)
ax2.xaxis.label.set_color('green')
ax2.tick_params(axis='x', colors='green')

ax4.plot(df['Apparent Water Resistivity'], df['DEPTH'], color='red', linewidth=1.25)
ax4.set_xscale('log')
ax4.set_xlim(0.2, 200)
ax4.xaxis.label.set_color('red')
ax4.tick_params(axis='x', colors='red')

ax3.plot(df['NPHI'], df['DEPTH'], color='red', linewidth=1.25)
ax3.set_xlim(0.6, 0)
ax3.xaxis.label.set_color('red')
ax3.tick_params(axis='x', colors='red')
ax3.set_xticks([0.60,0.4,0.2, 0])





# Manually set the y-axis for all subplots
for ax in [ax1, ax2, ax3, ax4]:
    #ax.set_ylim(df['DEPTH'].min(), df['DEPTH'].max())  # Ensure y-axis is correctly set
    ax.xaxis.set_ticks_position('top')
    ax.xaxis.set_label_position('top')
    # Define the number of gridlines (20) for each axis
    x_min, x_max = ax.get_xlim()

    if ax != ax4:  # Logarithmic scale for Resistivity plot
        grid_positions = np.linspace(x_min, x_max, 20)  # Linear-spaced gridlines

    elif ax == ax4:
        nothing = "True"



    if ax != ax4:
        ax.set_xticks(grid_positions, minor=True)  # Use minor ticks to avoid excessive labeling
        ax.grid(True, axis='x', linestyle='--', linewidth=0.5, which='minor')  # Gridlines on minor ticks
    elif ax == ax4:
        ax.grid(True, axis='x')   

# Use a for loop to assign x-axis labels from curve_names
axes_list = [ax1, ax2, ax3, ax4]  # List of axes
for i, ax in enumerate(axes_list):
    ax.set_xlabel(curve_names[i])  # Set the x-axis label from curve_names

# Hide y-tick labels for all axes except the first one
for ax in fig.axes[1:]:
    plt.setp(ax.get_yticklabels(), visible=False)


# Adjust spacing to prevent overlap
fig.subplots_adjust(wspace=.1, hspace=0)

# Title and finalize layout
#plt.suptitle('WR313 H001', x=0.1, horizontalalignment='left', fontsize='xx-large', fontweight='bold')

# Save plot

# Show plot
# plt.show()
plt.savefig(r'.\plots\HW_3_PartI_2_Well_logs_plot.png', format='png', dpi=300, bbox_inches='tight')
plt.close()
```
